```python
import requests
from rich.console import Console
from rich.panel import Panel
from rich.markdown import Markdown

console = Console()
chat_history = []

def get_llm_response_api(user_text):
    global chat_history
    default_system_prompt = (
        '''
        Ты генерал вернулся с фронта у тебя на руке часы. Ты находишься на пероне и к тебе подходит группа людей
        '''
    )
    if not chat_history:
        chat_history = [{"role": "system", "content": default_system_prompt}]
    chat_history.append({"role": "user", "content": user_text})
    headers = {
        "Authorization": "Bearer gpustack_ed6c3026e211c1e1_e301d483e74964cee0205629916d952d",
        "Content-Type": "application/json"
    }
    data = {
        "model": "qwen3-14b-awq",
        "messages": chat_history,
        "temperature": 0.8,
        "top_p": 0.95
    }
    response = requests.post("http://gojdaforever.ru:9696/v1/chat/completions", headers=headers, json=data, timeout=30)
    response.raise_for_status()
    result = response.json()
    msg = result['choices'][0]['message']
    content = msg.get('content') or msg.get('reasoning_content') or ""
    final_answer = content.strip()
    chat_history.append({"role": "assistant", "content": final_answer})
    return final_answer

def display_full_chat():
    for msg in chat_history:
        if msg["role"] == "system":
            console.print(Panel(Markdown(msg["content"]), title="system", border_style="yellow"))
        elif msg["role"] == "user":
            console.print(Panel(Markdown(msg["content"]), title="user", border_style="blue"))
        elif msg["role"] == "assistant":
            console.print(Panel(Markdown(msg["content"]), title="assistant", border_style="red"))

def display_last_messages():
    last_messages = chat_history[-2:] if len(chat_history) >= 2 else chat_history
    for msg in last_messages:
        if msg["role"] == "user":
            console.print(Panel(Markdown(msg["content"]), title="user", border_style="blue"))
        elif msg["role"] == "assistant":
            console.print(Panel(Markdown(msg["content"]), title="assistant", border_style="red"))

def chat(user_input: str):
    global chat_history
    if user_input == "/exit":
        console.print(Panel("Сеанс завершён.", title="system function", border_style="bright_red"))
        return False
    elif user_input == "/clear":
        system_msg = chat_history[0] if chat_history else {"role": "system", "content": "Ты генерал вернулся с фронта у тебя на руке часы. Ты находишься на пероне и к тебе подходит группа людей"}
        chat_history = [system_msg]
        console.print(Panel("История очищена", title="system function", border_style="yellow"))
        display_full_chat()
        return True
    elif user_input.startswith("/system "):
        new_prompt = user_input[8:].strip()
        if new_prompt:
            chat_history = [{"role": "system", "content": new_prompt}]
            console.print(Panel(f"Новый системный промпт: {new_prompt}", title="change system promt", border_style="yellow"))
        display_full_chat()
        return True
    reply = get_llm_response_api(user_input)
    display_last_messages()
    return True

if __name__ == "__main__":
    while True:
        inputMessage = console.input("[bold cyan]Введите сообщение:[/bold cyan] ")
        should_continue = chat(inputMessage)
        if not should_continue:
            break
```
